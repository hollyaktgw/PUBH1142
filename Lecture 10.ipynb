```python
import numpy as np
import seaborn as sns

# seed the numpy pseudo-random number generator
np.random.seed(28)

# Create an array of 1000 random values using a continquous uniform distribution between 50 and 100 and assign it to the variable population
population = np.random.uniform(50, 100, 1000)
```

```python
# make a histogram of the population with seaborn
sns.histplot(population, kde=True, color='blue', bins=30)

```

```python
# calculate the mean of the population
mean = np.mean(population)
print("Mean of the population is: ", mean)
```

```python
# Seed the numpy pseudo-random number generator
np.random.seed(28)

# select a random sample of 30 values from the population and assign it to the variable sample
sample = np.random.choice(population, 30)

# Calculate the mean of the sample
sample_mean = np.mean(sample)
print("Mean of the sample is: ", sample_mean)
```

```python
# create a for-loop that calculate the mean of 30 random samples from the population 1000 times and store the mean values in a list, then make a histogram of the list
means = []
for i in range(1000):
    sample = np.random.choice(population, 30)
    sample_mean = np.mean(sample)
    means.append(sample_mean)

sns.histplot(means, kde=True, color='red', bins=30) 
print("The mean of the means is: ", np.mean(means))
print("The standard deviation of the means is: ", np.std(means))
print("The standard error of the means is: ", np.std(means)/np.sqrt(1000))
```

```python
# how to select the 2.5th percentile 
print("The 2.5th percentile is: ", np.percentile(means, 2.5))
```

```python
# how to select the 97.5th percentile
print("The 97.5th percentile is: ", np.percentile(means, 97.5))
```
